```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt

G = ct.tf([0.4], [1., -1])
T = np.linspace(0, 10, 101)
T, yout = ct.forced_response(G, T=T,U=np.sin(T))
plt.plot(T, yout)
plt.plot(T, np.sin(T))
plt.title('forced response')
plt.xlabel('t')
plt.ylabel('yout')
plt.legend(['yout', 'sin(t)'])
plt.grid()
plt.show()
print(yout)
```
